```python
import pickle
import json
import pandas as pd
import re
```

```python
def concat_conversation(data):
    conversations = []
    summaries = []
    for i in range(0, len(data)):
        if len(data[i]['dialogue'].split('\r\n')) > 1:
            sentences = data[i]['dialogue'].split('\r\n')
            
        else:
            sentences = data[i]['dialogue'].split('\n')
            
        if len(sentences) == 1:
            continue
            
    
        summaries.append(data[i]['summary'].strip('\n').replace('\r\nt', ' '))

        conversations.append(" </s><s> ".join(sentences))
        
    return conversations, summaries
    

def transform_format(prefix):
    with open("./raw_data/"+prefix + '.json', encoding = 'utf8') as json_file:
        data = json.load(json_file)
        
    cons, sums = concat_conversation(data)
    
    
    print(len(cons))
    print(len(sums))
    
    
    data_dict = {'text':[], 'summary':[]}
    for i in range(0,len(cons)):
        data_dict['text'].append(cons[i])
        data_dict['summary'].append(sums[i])
    pd_data = pd.DataFrame.from_dict(data_dict)
    pd_data.to_csv(prefix+'.csv', index = False)
    
    
```

```python
for u in ['train', 'val', 'test']:
    transform_format(u)
```
